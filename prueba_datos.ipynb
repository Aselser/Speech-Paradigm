```python
import re

# Define the path to your log file
log_file_path = r'data\agus1130\paradigma.log'

# Initialize a list to store parsed log entries
log_entries = []

# Open and read the log file
with open(log_file_path, 'r', encoding='utf-8') as log_file:
    for line in log_file:
        # Example: Extract lines that contain 'EXP' entries
        if 'EXP' in line:
            log_entries.append(line.strip())

# Display the first few log entries
for entry in log_entries:  # Adjust the number to see more or fewer entries
    print(entry)
```
